```python
import pandas as pd
import numpy as np

import os

import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

import folium
```

# Data

```python
DATA_PATH = '/content/drive/MyDrive/my_ws/edu/data_analysis/apartment_transaction_price_analysis/data'
```

```python
dong_lat_lng_df = pd.read_csv(f'{DATA_PATH}/{CITY}/동별좌표.csv')
```

```python
apt_df = pd.read_csv(f'{DATA_PATH}/{CITY}/apartments_df.csv')
```

# Functions & Variables

```python
CITY = '서울특별시'
```

```python
CITIES = ['서울특별시', '부산광역시']
```

```python
CITY_LOC = np.mean(dong_lat_lng_df['lat']), np.mean(dong_lat_lng_df['lng'])
```

```python
def save_df(df, file_name):
  os.makedirs(f'{DATA_PATH}/{CITY}', exist_ok=True)
  df.to_csv(f'{DATA_PATH}/{CITY}/{file_name}.csv', index=False)
```

```python
def search_location(dong):
  try:
    dong_data = dong_lat_lng_df[dong_lat_lng_df['dong'] == dong]
    return (dong_data['lat'].values[0], dong_data['lng'].values[0])
  except:
    print(f'search_location({dong}) 실패')
    return (np.nan, np.nan)
```

```python
def draw_map(df, target_col, radius_scale_factor, map, color='blue'):
  for idx, row in df.iterrows():
    location=[row['lat'], row['lng']]
    folium.CircleMarker(
        location = location,
        popup = row['dong'] + str(int(row[target_col])),
        radius = row[target_col] / radius_scale_factor,
        color=color,
        fill = True
    ).add_to(map)
```

```python
def initialize_map(zoom_start=11):
  return folium.Map(location=CITY_LOC, zoom_start=zoom_start)
```

## 동별 집 값 평균 및 거래량

```python
dong_transaction_price_mean = apt_df.groupby(by='dong')['transaction_real_price'].mean().reset_index()
dong_transaction_price_mean
```

```python
dong_transaction_count = apt_df.groupby(by='dong')['transaction_real_price'].count().reset_index()
dong_transaction_count
```

```python
dong_transaction_analysis = pd.merge(dong_transaction_price_mean, dong_transaction_count, on='dong')
dong_transaction_analysis.columns = ['dong', 'mean_price', 'count']
dong_transaction_analysis.head()
```

```python
dong_transaction_analysis[['lat', 'lng']] = dong_transaction_analysis['dong'].apply(lambda x: pd.Series(search_location(x)))
dong_transaction_analysis
```

### scatterplot

#### 동별 평균 집값 분포

## 컬럼 추가

### transaction_year_month -> year, month 컬럼 추가

```python
np.unique(apt_df['transaction_year_month'].values)
```

'200812'와 '200901'은 한 달 차이지만 숫자로는 90 가까운 차이가 남 <br>
그래서 연도 컬럼과 달 컬럼으로 분리하고자 함

```python
apt_df['transaction_year'] = (apt_df['transaction_year_month']/100).astype(int)
apt_df['transaction_year']
```

```python
apt_df['transaction_month'] = (apt_df['transaction_year_month']%100)
apt_df['transaction_month']
```

```python
apt_df.head(1)
```

### transaction_date에서 첫번째 날짜 추출해서 컬럼으로 추가

- `11~20`, `1~10`과 같은 값으로 구성되어있는데 이를 11, 1의 값을 갖는 컬럼 추가

```python
np.unique(apt_df['transaction_date'].values)
```

```python
apt_df['transaction_date_first'] = apt_df['transaction_date'].str.split('~').str[0].astype(int)
apt_df.head(1)
```

### 거래가/m2 컬럼 추가

```python
apt_df['real_price_per_m2'] = apt_df['transaction_real_price'] / apt_df['exclusive_use_area']
display(apt_df.head(1))
len(apt_df)
```

### 동별 거래면적, 실거래가, 거래가/m2 평균 컬럼 추가
